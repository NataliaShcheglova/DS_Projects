## Исследуем модели

### Постановка задачи 

Необходимо построить систему, способную проанализировать поведение клиентов и предложить пользователям новый тариф: «Смарт» или «Ультра». Это **задача классификации.** 

В исходном датасете данные о поведении клиентов, которые уже перешли на эти тарифы. 
Необходимо построить модель для задачи классификации, которая выберет подходящий тариф.

Данные в столбце `is_ultra` говорят о том,  каким тарифом пользовался клиент в течение месяца («Ультра» — 1, «Смарт» — 0).Таким образом у нас два класса, с которми мы будем работать. 

### Создадим переменные для признаков 

Создадим из две переменные с признаками на основе обущающего датасета `data_train`, тестового `data_test` и валидационного `data_valid` датасетов:

- `features`  — признаки. Содержит все столбцы, кроме `is_ultra`.
- `target` — целевой признак. Соба запишем данные столбца `is_ultra`.

**Переменные для обучающего датасета:**

```python
features_train = data_train.drop(['is_ultra'], axis=1)
target_train = data_train['is_ultra']
print(features_train.shape)
print(target_train.shape)
```

**Переменные для валидационного датасета:**

```python
features_valid = data_valid.drop(['is_ultra'], axis=1)
target_valid = data_valid['is_ultra']
print(features_valid.shape)
print(target_valid.shape)
```

**Переменные для тестового датасета:**

```python
features_test = data_test.drop(['is_ultra'], axis=1)
target_test = data_test['is_ultra']
print(features_test.shape)
print(target_test.shape)
```

### Создадим и обучим модели 

#### Дерево решений 

Обучим модели с помощью дерева решений с глубиной дерева от 1 до 6. Для этого:
1. Создадим модель с помощью дерева решений, 
2. Обучим модель на тренировочной выборке с заданой глубиной дерева (от 1 до 6),
3. Проверим качекство моели на валидационной выборке и выберем лучшую модель. 
4. Сохраним модель с лучшим значением метрики accuracy на тренировочном (обучающем) датасете. 

Параметры — это настройки, которые помогают модели в работе. Параметры определяют, как предсказывает модель. Подберем наилучшую глубину дерева решений (параметр `max_depth`). 

Валидация помогает выявить переобучение модели. Валидация показывает, как поведёт себя модель с незнакомыми данными. Если правильность предсказаний резко падает, модель переобучена. 
Вберем наилучшее дерево с помощью проверки на валидационной выборке. 


```python
%%time 
# строка для подсчета времени расчета 
best_model_tree = None
best_result_tree = 0
# в цикле перебираем глубину дерева, чтоб выбрать лучшую модель 
for depth in range(1, 6):
    # создадим пустую модель дерева решений и сохраним ее в переменной с фиксированной псевдослучайностью random_state
    model_tree = DecisionTreeClassifier(random_state=12345, max_depth=depth) 
    # обучиим модель с заданной глубиной дерева
    model_tree.fit(features_train, target_train) 
    predictions_tree = model_tree.predict(features_valid) # получим предсказания модели с помощью валидационной выборки
    result_tree = accuracy_score(target_valid, predictions_tree) # посчитаем качество модели с помощью валидационной выборки
    if result_tree > best_result_tree:
        best_model_tree = model_tree
        best_result_tree = result_tree
        
print("Accuracy наилучшей модели на валидационной выборке:", best_result_tree)
print("Глубина дерева лучшей модели:", best_model_tree)
```

#### Случайный лес 

Алгоритм обучает большое количество независимых друг от друга деревьев, а потом принимает решение на основе голосования. Случайный лес помогает улучшить результат предсказания и избежать переобучения.

Обучим модели случайного леса с числом деревьев от 1 до 10. Для этого:
1. при инициализации модели указываем число деревьев равное состоянию счётчика циклов — est;
2. обучим модель на тренировочной выборке,
3. сохраним модель с лучшим значением accuracy на валидационной выборке.


```python
%%time
best_model_forest = None
best_result_forest = 0
# в цикле перебираем количество деревьев, чтоб выбрать лучшую модель
for est in range(1, 20):
    # в цикле перебираем глубину деревьев, чтоб выбрать лучшую модель
    for depth in range(1, 15):
        # обучим модель с заданным количеством деревьев
        model_forest = RandomForestClassifier(random_state=12345, 
                                              n_estimators=est,  
                                              max_depth=depth)
        # обучим модель на тренировочной выборке
        model_forest.fit(features_train, target_train)
        # посчитаем качество модели на валидационной выборке
        result_forest = model_forest.score(features_valid, target_valid)
        if result_forest > best_result_forest:
            best_model_forest = model_forest  # сохраним наилучшую модель
            # сохраним наилучшее значение метрики accuracy на валидационных данных
            best_result_forest = result_forest

print("Accuracy наилучшей модели на валидационной выборке:", best_result_forest)
print("Параметры лучшей модели:", best_model_forest)
```

Обучим модели случайного леса с числом деревьев от 10 до 50 с шагом 10. 

```python
%%time 
best_model_forest_v2 = None
best_result_forest_v2 = 0
# в цикле перебираем количество деревьев, чтоб выбрать лучшую модель 
for est in range(10, 200, 10):
    # в цикле перебираем глубину деревьев, чтоб выбрать лучшую модель 
    for depth in range(1, 15):
        model_forest_v2 = RandomForestClassifier(random_state=12345, 
                                                 n_estimators=est,  
                                                 max_depth=depth) # обучим модель с заданным количеством деревьев
        model_forest_v2.fit(features_train, target_train) # обучим модель на тренировочной выборке
        result_forest_v2 = model_forest_v2.score(features_valid, target_valid) # посчитаем качество модели на валидационной выборке
        if result_forest_v2 > best_result_forest_v2:
            best_model_forest_v2 = model_forest_v2 # сохраним наилучшую модель
            best_result_forest_v2 = result_forest_v2 # сохраним наилучшее значение метрики accuracy на валидационных данных

print("Accuracy наилучшей модели на валидационной выборке:", best_result_forest_v2)
print("Параметры лучшей модели:", best_model_forest_v2)
```

Сразу проверим Accuracy наилучшей модели на тестовой выборке для новой модели случайного леса:

```python
%%time 
result_forest_test_v2 = best_model_forest_v2.score(features_test, target_test)
print("Accuracy наилучшей модели на тестовой выборке:", result_forest_test_v2)
```

#### Логистическая регрессия 

Чтобы предсказать тариф, логистическая регрессия сначала считает, к какому классу (тарифу) близок клиент, затем в зависимости от ответа выбирает нужный класс (тариф). 

Обучим модель логистической регрессии:

```python
%%time
model_logistic = LogisticRegression() 
model_logistic.fit(features_train, target_train)
result_logistic = model_logistic.score(features_valid, target_valid)
print("Accuracy на валидационной выборке:", result_logistic)
```

## Проверим модели на адекватность

При создании модели машинного обучения можно использовать baseline модель. Она представляет собой простейшую (dummy) модель, которая всегда предсказывает наиболее часто встречающийся класс. Так мы получим контрольные данные для сравнительной оценки построенной нами более сложной модели и сможем быть уверены в качестве ее работы, например, в том, что она выдает не просто набор случайно подобранных данных. 

В библиотеке scikit-learn есть DummyClassifier() для задач классификации. 

Для проверки адексатности модели воспользуемся DummyClassifier из библиотеки sklearn, обратившись к документации по ссылке: https://scikit-learn.org/stable/modules/generated/sklearn.dummy.DummyClassifier.html

```python
%%time
#построение модели через DummyClassifier
from sklearn.dummy import DummyClassifier
model_dummy = DummyClassifier(strategy='most_frequent', random_state=2)
model_dummy.fit(features_train, target_train)
predict= model_dummy.predict(features_test)
print("Accuracy dummy модели на тестовой выборке:", model_dummy.score(predict, target_test))
```

**ВЫВОД:** 

Произведена проверка на адекватность с помощью DummyClassifier() библиотеки scikit-learn. 
Accuracy модели, созданной с помощью DummyClassifier ниже, чем у модели Случайный лес, следовательно модель адекватна.
