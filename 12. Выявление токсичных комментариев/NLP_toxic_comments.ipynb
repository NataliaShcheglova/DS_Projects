### Подготовим данные для обучения моделей

#### Лемматизация и очистка текста 

Прежде чем извлечь признаки из текста, упростим его.

**Создадим корпус постов.** 
Преобразуем столбец 'text' в список текстов. Переведём тексты в стандартный для Python формат: кодировку Unicode U. 

```python
corpus = list(df['text'])
```

Лемматизация (lemmatization) — приведение слова к начальной форме — лемме. 

**Напишем функцию lemmatize(text)**. На вход она принимает текст из корпуса и возвращает лемматизированную строку. 

```python
# Initialize spacy 'en' model, keeping only tagger component needed for lemmatization
nlp = spacy.load("en_core_web_sm", disable=['parser', 'ner'])

def lemmatize(text):
    doc = nlp(text)
    lemm_text = " ".join([token.lemma_ for token in doc]).lower()

    return lemm_text
```

**Напишем функцию clear_text(text)**, которая оставит в тексте только латинские символы и пробелы, дополнительно уберем лишние пробелы. На вход она принимает текст, а возвращает очищенный текст. 

```python
def clear_text(text):
    clear_text = re.sub(r'[^a-zA-Z]', ' ', text)
    clear_text = " ".join(clear_text.split())

    return clear_text
```

```python
print("Очищенный и лемматизированный текст:", lemmatize(clear_text(corpus[0])))
```

*Чтобы сэкономить время работы над проектом, сохраним лемматизированнй текст в отдельный файл, из которго будем его выгружать в дальнейшем, чтобы не проделывать лемматизацию каждй раз при перезапуске ядра после очереднх изменений. Если на какой-то машине код запускается впервые, то лемматизация текста и запись в файл будет проведена.*

```python
try:
    df = pd.read_csv('toxic_comments_lemma.csv')
except:
    df['lemma_text'] = df['text'].apply(lambda x: lemmatize(clear_text(x)))
    df.to_csv('toxic_comments_lemma.csv')
    df = pd.read_csv('toxic_comments_lemma.csv')
```

```python
df = df.drop(['Unnamed: 0'], axis=1)
df.tail()
```

### Подготовим выборки для обучения моделей

#### Создадим переменные для признаков

Создадим из две переменные с признаками и целевым признаком:

- target — целевой признак. Сюда запишем данные столбца `toxic`.

#### Разбивка данных на выборки 

Разобьем датасет на обучающую (80%) и тестовую (20%) выборки.

```python
features_train, features_test, target_train, target_test = train_test_split(features, target,
                                                                            test_size=0.2,
                                                                            random_state=STATE,
                                                                            stratify=target)
print('Обучающая выборка - признаки', features_train.shape)
print('Обучающая выборка - целевой признак', target_train.shape)
print('Тестовая выборка - признаки', features_test.shape)
print('Тестовая выборка - целевой признак', target_test.shape)
```

#### Вычислим TF-IDF для корпуса текстов
Функцию fit() будем запускать только на обучающей выборке. Иначе тестирование будет нечестным: в модели будут учтены частоты слов из тестовой выборки.

Создадим корпус текстов для обучающей выборки.

```python
corpus = features_train['lemma_text'].apply(lambda x: np.str_(x))
```

Найдём и исключим англоязычные стоп-слова (слова без смысловой нагрузки), предварительно загрузив список стоп-слов.  

```python
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
```

Создадим и обучим векторизатор TF-IDF

```python
count_tf_idf = TfidfVectorizer(stop_words=list(stop_words))
features_train_tfidf = count_tf_idf.fit_transform(corpus)

print("Размер матрицы:", features_train_tfidf.shape)
```

**ВЫВОД:**
- Проведена лемматизация и очистка текста.
- созданы переменные для признаков и целевого признака.
- Датасет разделен на обучающую и тестовую выборки в соотношении 80% на 20%. 
- Устранен стоп-слова.
- Создан и обучен векторизатор TF-IDF 

## Обучение

### Подготовка к обучению 


**Зададим параметры для моделей**

```python
param_grid_lr = {'model__C': [0.001, 0.01, 0.1, 1, 10],
                 'model__penalty': ['l1', 'l2']}

param_grid_lgbm = {
    'LGBM__num_leaves': [10, 20, 31, 40, 50],
    'LGBM__learning_rate': [0.05, 0.1],
    'LGBM__n_estimators': [10, 20]}
```

**Создадим пайплайн для моделей** 

```python
pipeline_lr = Pipeline([
    ('TF-IDF', TfidfVectorizer(stop_words=list(stop_words))),
    ('model', LogisticRegression(max_iter=500))])

pipeline_lgbm = Pipeline([
    ('TF-IDF', TfidfVectorizer(stop_words=list(stop_words))),
    ('LGBM', LGBMClassifier(class_weight = 'balanced', random_state=STATE))])
```

### Обучение моделей

#### LogisticRegression()

```python
%%time
grid_lr = GridSearchCV(pipeline_lr,
                        param_grid=param_grid_lr,
                        cv=5,
                        scoring='f1',
                        n_jobs=-1)

grid_lr.fit(corpus, target_train)

print('Параметр лучшей модели:', grid_lr.best_params_)
print('f1-мера лучшей модели: ', grid_lr.best_score_)
```

#### LGBMClassifier()

```python
%%time
grid_lgbm = GridSearchCV(pipeline_lgbm,
                        param_grid=param_grid_lgbm,
                        cv=5,
                        scoring='f1',
                        n_jobs=-1)

grid_lgbm.fit(corpus, target_train)

print('Параметр лучшей модели:', grid_lgbm.best_params_)
print('f1-мера лучшей модели: ', grid_lgbm.best_score_)
```

- LogisticRegression показала себя лечше (f1-мера = 0.781), значение f1-меры выше порогового значения 0.75. Кроме того, Логистическая регрессия намного быстрее, что является дополнительным бонусом. 
- LGBMClassifier (f1-мера = 0.697) не прошел пороговое значение. 

### Тестирование лучшей модели 

Обучим лучшую модель - LogisticRegression с лучшими параметрами, подобранными с помощью грид серча. Сделаем предсказания лучшей модели на тестовой выборке. Рассчитаем f1-меру лучшей модели на тестовой выборке.

```python
corpus_test = features_test['lemma_text'].apply(lambda x: np.str_(x))
features_test_tfidf = count_tf_idf.transform(corpus_test)
```

```python
%%time 
best_model = LogisticRegression(C=10, penalty='l2').fit(features_train_tfidf, target_train)
best_model_predict = best_model.predict(features_test_tfidf)

print ('f1-мера лучшей модели на тесте: ', f1_score(target_test, best_model_predict))
```

Значение метрики F1 лучшей модели на тесте равно 0.78,  что больше порогового значения 0.75. Модель прошла проверку.

**ВЫВОД:** 
- Обучены модели LogisticRegression() и LGBMClassifier().
- Подбор параметров осуществлен с помощью GridSearchCV.
- LogisticRegression показала себя лечше (f1-мера = 0.781), значение f1-меры выше порогового значения 0.75. Кроме того, Логистическая регрессия намного быстрее, что является дополнительным бонусом. 
- LGBMClassifier (f1-мера = 0.697) не прошел пороговое значение.  
- Проведено тестирование лучшей модели. Значение метрики F1 лучшей модели на тесте равно 0.78, что больше порогового значения 0.75. Модель прошла проверку.

Прежде чем перевести тексты в векторы, подготовим их. У BERT есть собственный токенизатор. Это инструмент, который разбивает и преобразует исходные тексты в список токенов, которые есть в словаре BERT. Лемматизация не требуется.

**Загрузим предобученную модель для токсичных англоязычных комментариев и инициализируем токенизатор**

```python
tokenizer = transformers.BertTokenizer.from_pretrained('unitary/toxic-bert')
model = transformers.BertModel.from_pretrained('unitary/toxic-bert')
```

**Проведем токенизацию выборки и преобразуем текст в номера токенов из словаря** методом encode().
Для корректной работы модели мы указали аргумент add_special_tokens (англ. «добавить специальные токены»), равный True. Это значит, что к любому преобразуемому тексту добавляется токен начала (101) и токен конца текста (102)

```python
tokenized = data_sample['text'].apply(
    (lambda x: tokenizer.encode(x[:512], add_special_tokens=True)))
```

Применим метод padding, чтобы после токенизации длины исходных текстов в корпусе были равными. Только при таком условии будет работать модель BERT. Пусть стандартной длиной вектора n будет длина наибольшего во всём датасете вектора. Остальные векторы дополним нулями. 

```python
max_len = 0
for i in tokenized.values:
    if len(i) > max_len:
        max_len = len(i)

padded = np.array([i + [0]*(max_len-len(i)) for i in tokenized.values])
```

Теперь поясним модели, что нули не несут значимой информации. Это нужно для компоненты модели, которая называется attention. Отбросим эти токены и «создадим маску» для действительно важных токенов, то есть укажем нулевые и не нулевые значения:

```python
attention_mask = np.where(padded != 0, 1, 0)
print(attention_mask.shape) 
```

Создание тензора из массива padded, что позволяет представить текстовые данные в виде числовых идентификаторов (токенов), которые модель может понимать. 

```python
%%time
input_ids = torch.tensor(padded)  
attention_mask = torch.tensor(attention_mask)

with torch.no_grad():
    last_hidden_states = model(input_ids, attention_mask=attention_mask)
```

### Подготовим выборки для обучения моделей
**Создадим переменные для признаков**

Создадим из две переменные с признаками и целевым признаком:

- features — признаки. Соберём все эмбеддинги в матрицу признаков вызовом функции concatenate().
- target — целевой признак. Сюда запишем данные столбца `toxic`.

```python
features_bert = last_hidden_states[0][:,0,:].numpy()
target_bert = data_sample['toxic']
```

**Разбивка данных на выборки**

Разобьем датасет на обучающую (80%) и тестовую (20%) выборки.

```python
features_train_bert, features_test_bert, target_train_bert, target_test_bert = train_test_split(features_bert, 
                                                                                                target_bert, 
                                                                                                stratify=target_bert,  
                                                                                                test_size=0.2, 
                                                                                                random_state=STATE)

print('Обучающая выборка - признаки', features_train_bert.shape)
print('Обучающая выборка - целевой признак', target_train_bert.shape)
print('Тестовая выборка - признаки', features_test_bert.shape)
print('Тестовая выборка - целевой признак', target_test_bert.shape)
```

### Обучение и тестирование модели

Обучим Логистическую регрессию, поскольку в первой части проекта она показала себя довольно хорошо. Кроме того, логистическая регрессия считается гораздо быстрее более сложных моделей. Подбор параметров проведем с помощью GridSearchCV с параметрами, заданными выше в проекте. 

Обучим модель с лучшими данными и проверими ее на тесте. 

```python
%%time 
best_model_bert = LogisticRegression(C=10, penalty='l2').fit(features_train_bert, target_train_bert)
bert_predict = best_model_bert.predict(features_test_bert)

print ('f1-мера лучшей модели на тесте: ', f1_score(target_test_bert, bert_predict))
```
