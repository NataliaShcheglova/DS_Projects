**Технологический процесс**
- `Rougher feed` — исходное сырье
- `Rougher additions` (или reagent additions) — флотационные реагенты: 
 - `Xanthate`— ксантогенат (промотер, или активатор флотации);
 - `Sulphate` — сульфат (на данном производстве сульфид натрия);
 - `Depressant` — депрессант (силикат натрия).
- `Rougher process` (англ. «грубый процесс») — флотация
- `Rougher tails` — отвальные хвосты
- `Float banks` — флотационная установка
- `Cleaner process` — очистка
- `Rougher Au` — черновой концентрат золота
- `Final Au` — финальный концентрат золота

**Параметры этапов**
- `air amount` — объём воздуха
- `fluid levels` — уровень жидкости
- `feed size`  — размер гранул сырья
- `feed rate` — скорость подачи

Наименование признаков в формате:
`[этап].[тип_параметра].[название_параметра]`

- `rougher` — флотация
- `primary_cleaner` — первичная очистка
- `secondary_cleaner` — вторичная очистка
- `final` — финальные характеристики

**Значения для блока `[тип_параметра]`:**
- `input` — параметры сырья
- `output` — параметры продукта
- `state` — параметры, характеризующие текущее состояние этапа
- `calculation` — расчётные характеристики


# Технологический процесс

Как золото получают из руды: стадии процесса.
Когда добытая руда проходит первичную обработку, получается дроблёная смесь. Её отправляют на флотацию (обогащение) и двухэтапную очистку.

**Опишем каждую стадию:**
1. **Флотация**

Во флотационную установку подаётся смесь золотосодержащей руды. После обогащения получается черновой концентрат и «отвальные хвосты», то есть остатки продукта с низкой концентрацией ценных металлов.
На стабильность этого процесса влияет непостоянное и неоптимальное физико-химическое состояние флотационной пульпы (смеси твёрдых частиц и жидкости).

2. **Очистка** 

Черновой концентрат проходит две очистки. На выходе получается финальный концентрат и новые отвальные хвосты.

- C — доля золота в концентрате после флотации/очистки - столбец  `rougher.output.concentrate_au`, переменная `c_output`;
- F — доля золота в сырье/концентрате до флотации/очистки - `rougher.input.feed_au`, переменная`f_input`;
- T — доля золота в отвальных хвостах после флотации/очистки - `rougher.output.tail_au`, переменная `t_tail`.

**Напишем функцию для расчета эффективности обогащения:**


```python
def recovery_check(line):

    c_output = line["rougher.output.concentrate_au"]
    f_input = line["rougher.input.feed_au"]
    t_tail = line["rougher.output.tail_au"]

    recovery = ((c_output * (f_input - t_tail)) /
                (f_input * (c_output - t_tail))) * 100

    return recovery
```

**Найдем MAE между своими расчётами и значением признака.** 

Проверку необходимо провести без учета нулевых значений, потому удалим строки с нулями из обучающей выборки и сохраним полученный датафрейм в отдельную переменную.

```python
gr_train_check = gr_train.dropna(
    subset=["rougher.input.feed_au", "rougher.output.concentrate_au",
            "rougher.output.tail_au", "rougher.output.recovery"],
    axis=0)
```

Найдем MAE (Средняя абсолютная ошибка) функцией mean_absolute_error()

```python
print("Значение MAE:",
      mean_absolute_error(gr_train_check["rougher.output.recovery"], gr_train_check.apply(recovery_check, axis=1)))
```

**ВЫВОД:**
Средняя абсолютная ошибка очень маленькая (значение MAE = 9.7x10 в степени-15), следовательно значения rougher.output.recovery в датафрейме посчитаны верно.

```python
def drop_zeroes(frame):
    for st in stage:
        for col in st:
            try:
                frame = frame.drop(index=frame[frame[col] == 0].index, axis=0)
            except:
                print('Нет колонки в тестовом датасете', col)
    return frame
```

## Метрика качества

Для решения задачи введём новую метрику качества — sMAPE (англ. Symmetric Mean Absolute Percentage Error, «симметричное среднее абсолютное процентное отклонение»).
Она похожа на MAE, но выражается не в абсолютных величинах, а в относительных. Почему симметричная? Она одинаково учитывает масштаб и целевого признака, и предсказания.

Метрика sMAPE вычисляется так:

где:

- **yi** - *Значение целевого признака для объекта с порядковым номером i в выборке, на которой измеряется качество*   (действительное значение целевого признака - `target`)
 
- **^yi** - *Значение предсказания для объекта с порядковым номером i, например, в тестовой выборке* (предсказанное значение целевого признака `prediction`);
 
- **N** - *Количество объектов в выборке* (количество значений признака - `len(target)`). 



Итоговая метрика складывается из двух величин:

### Напишем функцию для вычисления итоговой sMAPE

Срачала напишем функцию для расчета sMAPE:

```python
def s_mape_calculation(target, prediction):
    s_mape = (100/len(target)) * np.sum(2 * np.abs(prediction -
                                                   target) / (np.abs(target) + np.abs(prediction)))
    return s_mape
```

Напишем функцию для расчета Итогового sMAPE:

```python
def final_s_mape_calculation(rougher, final):
    final_s_mape = 0.25*rougher + 0.75*final
    return final_s_mape
```

Для оценки моделей создадим метрику с помощью make_scorer:

```python
score_check = make_scorer(s_mape_calculation, greater_is_better=False)
```

### GridSearchCV

Чтобы оценить гиперпараметры модели, используем GridSearchCV на данных обучающей выборки.
GridSearchCV уже включает в себя кросс-валидацию. 

*Проверила значения, которые получаем с помощью cross_val_score() - результат хуже, чем при использовании GridSearchCV, потому закомментила эту часть кода, чтоб не тратить время на расчет при перезапуске ядра**  

**Будем использовать модели LinearRegression(), DecisionTreeRegressor(), RandomForestRegressor(), GradientBoostingRegressor(), а затем сравним с DummyRegressor()**. 

Создадим функцию поиска наилучших параметров моделей и лучшего sMAPE с помощью кроссвалидации в составе GridSearchCV.

```python
def grid_search_calc(model_cv, params, features, target):
    
    grid_search = GridSearchCV(model_cv,
                      param_grid=params,
                      cv = 5,
                      scoring = score_check,
                      n_jobs = -1,  
                      verbose=10
                    )
    
    grid_search .fit(features, target)
    
    return grid_search.best_params_, grid_search.best_score_
```

Обозначим гиперпараметры, которые будут перебираться в grid_search.

**1. Изучено, как меняется концентрация металлов (Au, Ag, Pb) на различных этапах очистки**

- Концентрация серебра увеличивается после флотации, затем понижается с после каждого этапа очистки.
- Концентрация свинца повышается до первичной очистки, а после фторичной очистки остается практически без изменений.

То, что какое-то количество металлов остается, можно обяснить особенностями техпроцесса. 

**2. Изучены и удалены выбросы в обучающей выборке и исходном датасете** 
- Гистограммы и описательная статистика позволили вычвить выбросы (нулевые значения концентраций). 
 - Мы видим, что на начальной стадии (до флотации) отсутствуют вбросы (значения, равные нулю), но в ходе технологического процесса при переходе от стадии к стадии количество выбросов резко возрастает и затем снижается.
 - Предположу, что в процессе очистки смеси золотой руды какое-то количество золота также уносится из сырья вместе с продуктами очистки, в результате в черновом концентрате (после флотации) поялвяются выбросы в данных, а затем после первичной и вторичной очистки гранулы, не содержащие золото, тоже уносятся, и количество выбросов снижается. Но процент выбросов настолько мизерный на всех стадиях, что не влияет на результат техпроцесса.
- Удалены выбросы в исходной и обучающей выборке.  Количество выбросов незначительное, тем не менее они могут негативно повлиять на обучение модели. В тестовой выборке нет данных по концентрации металлов, как мы выяснили ранее в проекте, потому удалять выбросы не требуется.
