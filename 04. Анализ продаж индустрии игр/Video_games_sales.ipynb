- Разные платформы существовали в разные периоды времени, и срок их работы тоже отличается. 

#### Выясним, за какой характерный срок появляются новые и исчезают старые платформы

```python
# создадим талицу по платформам с учетом времени появления и угасания продаж, запишем ее в переменную life_cycle
life_cycle = data.pivot_table(index='platform',values='year_of_release',aggfunc=['max','min'])
life_cycle.columns = ['max','min']
# рассчитаем срок жизни каждой платформ и запишем в таблицу 
life_cycle['delta'] = life_cycle['max'] - life_cycle['min']
# отсортируем по убыванию по максимальному значению 
life_cycle = life_cycle.sort_values('max',ascending=False)
life_cycle 
```

**Посчитаем среднее значение и медиану для срока жизни платформ** 

```python
life_cycle_mean = life_cycle['delta'].mean()
life_cycle_median = life_cycle['delta'].median()
print('Средний срок существования игровой пратформы', life_cycle_mean, 'лет')
print('Медианный срок существования игровой пратформы', life_cycle_median, 'лет')
```

**Для определения продолжительности жизни платформы будем использовать медианное значение, поскольку среднее будет искажено платформами-длогожителями (например, DS - 28 лет, PC - 31 год). Медиана даст более точное представление**

## Проверка гипотез

Для гипотез ниже будеи использовать **t-тест**, поскольку в обоих случаях рассматриваем гипотезу о равенстве средних двух генеральных совокупностей, генеральные совокупности не зависят друг от друга и выборочне средние нормально распределены. 


 В качестве уровня значимости alpha возьмем 0.05 (5%).

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

**H_0:** Средний пользовательский рейтинг (user_score) платформы "Xbox One" = среднему прльзовательскому рейтингу (user_score) платформы "PC"
**H_a:** Средний пользовательский рейтинг (user_score) платформы "Xbox One" ≠ среднему прльзовательскому рейтингу (user_score) платформы "PC"


```python
# alpha = задаем значение уровня значимости
alpha = .05
# results = вызов метода для проверки гипотезы
results = st.ttest_ind(
    actual_term[actual_term['platform'] =='XOne']['user_score'].dropna(), 
    actual_term[actual_term['platform'] =='PC']['user_score'].dropna(),
    equal_var = False)

# вывод значения p-value на экран 
print('p-значение:', results.pvalue)
# условный оператор с выводом строки с ответом
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

**Вывод: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.** Поскольку нулевую гипотезу, которая была задана, отвергнуть не удалось. 

### Средние пользовательские рейтинги жанров Action и Sports разные

**H_0:** Средний пользовательский рейтинг (user_score) жанра "Action" = среднему прльзовательскому рейтингу (user_score) жанра "Sports" 

**H_a:** Средний пользовательский рейтинг (user_score) жанра "Action" ≠ среднему прльзовательскому рейтингу (user_score) жанра "Sports"




```python
# alpha = задаем значение уровня значимости
alpha = .05
# results = вызов метода для проверки гипотезы
results = st.ttest_ind(
    actual_term[actual_term['genre'] =='Action']['user_score'].dropna(), 
    actual_term[actual_term['genre'] =='Sports']['user_score'].dropna(),
    equal_var = False)

# вывод значения p-value на экран 
print('p-значение:', results.pvalue)
# условный оператор с выводом строки с ответом
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

**Вывод: Средние пользовательские рейтинги жанров Action и Sports разные**. Поскольку в результате t-теста мы отвергли заданную нулевую гипотезу.
